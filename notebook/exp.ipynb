```python
from hfselect import Dataset, compute_task_ranking

# Load target dataset from the Hugging Face Hub
# NusaX-Senti requires a subset (language code), e.g. "indonesian"
dataset = Dataset.from_hugging_face(
    name="indonlp/NusaX-senti",
    subset="indonesian",
    split="train",
    text_col="text",
    label_col="label",
    is_regression=False,
    num_examples=500,
    seed=42
)

# Fetch ESMs and rank tasks
task_ranking = compute_task_ranking(
    dataset=dataset,
    model_name="LazarusNLP/NusaBERT-base-SmSA"
)

# Display top 5 recommendations
print(task_ranking[:5])
```
